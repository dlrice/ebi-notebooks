```python
import requests
from collections import defaultdict
```

```python
def get_deep_keys(o, depth=0):
    if type(o) == dict:
        for key, value in o.items():
            space = '  '*depth
            print(f'{space}{key}')
            get_deep_keys(value, depth+1)
    elif type(o) == list:
        for el in o:
            get_deep_keys(el, depth+1)
            
def show_all_comment_type_entries(o):
    if type(o) == dict:
        if 'commentType' in o:
            k = tuple(o.keys())
            v = o['commentType']
            D[v].add(k)
        else:
            for key, value in o.items():
                show_all_comment_type_entries(value)
    elif type(o) == list:
        for el in o:
            show_all_comment_type_entries(el)
```

```python
accession = 'P05067'
r = requests.get(f'http://wwwdev.ebi.ac.uk/uniprot/api/uniprotkb/accession/{accession}', headers={"Accept":"application/json"})
d = r.json()
D = defaultdict(set)
show_all_comment_type_entries(d)
```

```python
print(f'*** All commentType properties for accession {accession} ***', end='\n'*2)
for k in sorted(D.keys()):
    v = D[k]
    print(k)
    required = set.intersection(*[set(w) for w in v])
    union = set([y for x in v for y in x])
    for el in sorted(union):
        if el in required:
            r = ''
        else:
            r = ' (optional)'
        print(f'  |{el}{r}')
    print()
```
